## 02. Data Wrangling

```python
df_customer_value.columns
```

```python
# Check for uniqueness
# Convert list of dictionaries to DataFrame
df = pd.DataFrame(df_customer_value)
```

```python
# Check for unique values in each column
non_unique_columns = df.columns[df.nunique() == 1]

# Print non-unique column names
if len(non_unique_columns) > 0:
    print("The values in the following column(s) are not unique:", non_unique_columns.tolist())
else:
    print("All values in the dataset are unique.")
```

```python
# check for mixed types
for col in df_customer_value.columns.tolist():
  weird = (df_customer_value[[col]].applymap(type) != df_customer_value[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customer_value[weird]) > 0:
    print (col)
```

```python
# Droping unwanted columns
df_customer_value=df_customer_value.drop(columns=['Education'])
```
